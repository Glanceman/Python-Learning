Load libraries

```python
import pandas as pd
from sklearn import tree,datasets,preprocessing
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from scipy.io import arff
import graphviz
import numpy as np
```

load dataset

```python
RawData=arff.loadarff("./Dataset/vertebral_column_data/column_3C_weka.arff")
#RawData = np.loadtxt("./Dataset/vertebral_column_data/column_3C.dat", delimiter=' ')
print(RawData[0])
```

```python
Data = pd.DataFrame(RawData[0])
FeatureName=Data.columns
Data.iloc[:,6]=Data.iloc[:,6].str.decode('utf-8')
print(Data.iloc[0:2,:])
print("------------------------------------------")
print(FeatureName)
```

Assign the data and target

```python
XData=np.array(Data.iloc[:,0:5])
YLabels=np.array(Data.iloc[:,6])
LabelEncoder= preprocessing.LabelEncoder()
YTarget=LabelEncoder.fit_transform(YLabels)
ClassNames=LabelEncoder.classes_
print(XData[0],YTarget[0])
print("ClassName: ", ClassNames)
```

Spilt the data 7:3

```python
XTrain73, XTest73,YTrain73, YTest73 = train_test_split(XData,YTarget,test_size=0.3,random_state=1,stratify=YTarget)
print(len(YTrain73)/len(XData))
print(YTrain73)
```

construct decision tree

```python
DecisionTree73=tree.DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=1)
DecisionTree73=DecisionTree73.fit(XTrain73,YTrain73)
# YPredict73 = DecisionTree73.predict(XTest73)
# print("Tree(7 train , 3 test): ",accuracy_score(YTest73,YPredict73))
```

Visualiztion Decision Tree

```python
dot_data73 = tree.export_graphviz(DecisionTree73,out_file=None,feature_names=FeatureName[0:5],class_names=ClassNames, filled=True)
graph=graphviz.Source(dot_data73)
graph.render("Vertebral Column73",view=True)
```

construct tree in 9:1

```python
XTrain91, XTest91,YTrain91, YTest91 = train_test_split(XData,YTarget,test_size=0.1,random_state=1,stratify=YTarget)
print(len(YTrain91)/len(XData))
DecisionTree91=tree.DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=1)
DecisionTree91=DecisionTree91.fit(XTrain91,YTrain91)
dot_data91 = tree.export_graphviz(DecisionTree91,out_file=None,feature_names=FeatureName[0:5],class_names=ClassNames, filled=True)
graph=graphviz.Source(dot_data91)
graph.render("Vertebral Column91",view=True)
```

construct Decision Tree into 5:5

```python
XTrain55, XTest55,YTrain55, YTest55 = train_test_split(XData,YTarget,test_size=0.5,random_state=1,stratify=YTarget)
print(len(YTrain55)/len(XData))
DecisionTree55=tree.DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=1)
DecisionTree55=DecisionTree55.fit(XTrain55,YTrain55)
dot_data55 = tree.export_graphviz(DecisionTree55,out_file=None,feature_names=FeatureName[0:5],class_names=ClassNames, filled=True)
graph=graphviz.Source(dot_data55)
graph.render("Vertebral Column55",view=True)
```

construct Decision Tree into 3:7

```python
XTrain37, XTest37,YTrain37, YTest37 = train_test_split(XData,YTarget,test_size=0.7,random_state=1,stratify=YTarget)
print(len(YTrain37)/len(XData))
DecisionTree37=tree.DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=1)
DecisionTree37=DecisionTree37.fit(XTrain37,YTrain37)
dot_data37 = tree.export_graphviz(DecisionTree37,out_file=None,feature_names=FeatureName[0:5],class_names=ClassNames, filled=True)
graph=graphviz.Source(dot_data37)
graph.render("Vertebral Column37",view=True)
```

Contruct tree in 1:9

```python
XTrain19, XTest19,YTrain19, YTest19 = train_test_split(XData,YTarget,test_size=0.9,random_state=1,stratify=YTarget)
print(len(YTrain19)/len(XData))
DecisionTree19=tree.DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=1)
DecisionTree19=DecisionTree19.fit(XTrain19,YTrain19)
dot_data19 = tree.export_graphviz(DecisionTree19,out_file=None,feature_names=FeatureName[0:5],class_names=ClassNames, filled=True)
graph=graphviz.Source(dot_data19)
graph.render("Vertebral Column19",view=True)
```

Score

```python
YPredict91 = DecisionTree91.predict(XTest91)
YPredict73 = DecisionTree73.predict(XTest73)
YPredict55 = DecisionTree55.predict(XTest55)
YPredict37 = DecisionTree37.predict(XTest37)
YPredict19 = DecisionTree19.predict(XTest19)
print("Accurary Score")
print("Tree(9 train , 1 test): ",accuracy_score(YTest91,YPredict91))
print("Tree(7 train , 3 test): ",accuracy_score(YTest73,YPredict73))
print("Tree(5 train , 5 test): ",accuracy_score(YTest55,YPredict55))
print("Tree(3 train , 7 test): ",accuracy_score(YTest37,YPredict37))
print("Tree(1 train , 9 test): ",accuracy_score(YTest19,YPredict19))

print("confusion matrix")
print("Tree(9 train , 1 test): ")
print(confusion_matrix(YTest91,YPredict91))
print("---------------------------------------------")
print("Tree(7 train , 3 test): ")
print(confusion_matrix(YTest73,YPredict73))
print("---------------------------------------------")
print("Tree(5 train , 5 test): ")
print(confusion_matrix(YTest55,YPredict55))
print("---------------------------------------------")
print("Tree(3 train , 7 test): ")
print(confusion_matrix(YTest37,YPredict37))
print("---------------------------------------------")
print("Tree(1 train , 9 test): ")
print(confusion_matrix(YTest19,YPredict19))
```
