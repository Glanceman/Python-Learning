```python
#import lib
import requests as R
import bs4 as BS

targetUrl = "https://www.bbc.com/news"
res = R.get(targetUrl)
html = res.text
bs = BS.BeautifulSoup(html)
#print(bs.prettify())

# Get all new link 
linkTags = bs.find_all("a")
candidateLinks=[]
for linkTag in linkTags:
    href = linkTag["href"]
    if href is not None:
        if href.startswith("/news/"):
            candidateLinks.append("https://www.bbc.com"+href)
print(candidateLinks[:10])

# Retrieve data from the candidate links
for link in candidateLinks[:20]:
    res = R.get(link)
    bs = BS.BeautifulSoup(res.text)
    # display information
    print(bs.title.text)
    paras = bs.find_all("p")
    text = [para.text for para in paras]
    print(text)
    print("-------------------------------------------")
    
```

```python
# import lib
import xml.etree.ElementTree as ET
from io import open
import re
# load xml file 
def Xml_To_Text(path:str)->str:
    tree:ET= ET.parse(path)
    root:ET.Element = tree.getroot()
    text:str = ET.tostring(root, encoding='utf-8', method='text').decode('utf-8')
    y= re.sub(r"( ){2,}","",text)
    y= re.sub(r"\n+","\n",y)
    text = y
    return text

xmlPath:str = "./sample-src.xml"
res = Xml_To_Text(xmlPath)
print(res)
with open('./OutputFile/FromXMLText.txt', 'w', encoding="utf-8") as file:
    # Write the text content to the file
    file.write(res)
```

## 1
user id (Integer;1-200) and age(Integer;18-100): Histogram 
height (Float;100.0-200.0) and weight (Float;30.0-100.0): scatter
sex (Binary;Male/Female) : pie chart

```python
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2
user_id:np.ndarray =np.empty(shape=(0,1))
age:np.ndarray = np.empty(shape=(0,1))
sex:np.ndarray = np.empty(shape=(0,1))
height:np.ndarray = np.empty(shape=(0,1))
weight:np.ndarray= np.empty(shape=(0,1))
for i in range(0,200):
    user_id= np.append(user_id,i)
    sex = np.append(sex, "M" if random.random()>0.5 else "F")
    height = np.append(height, random.random()*100+100)
    weight=np.append(weight,random.random()*70+30)
    age = np.append(age,round(random.random()*82+18))

print("{} : {} : {} : {} : {}".format(user_id,age,sex,height,weight))

#plot histogram
plt.title("Age")
plt.hist(age, bins=20)
plt.show()

# plot height, weight
plt.scatter(weight,height)
plt.title("Weight vs Height")
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()

# plot sex
plt.pie(np.array([np.count_nonzero(sex == "M"),np.count_nonzero(sex == "F")]),labels=["Male","Female"])
plt.title("Sex")
plt.legend()
plt.show()
```
