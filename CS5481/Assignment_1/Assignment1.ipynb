```python
#import lib
import requests as R
import bs4 as BS

targetUrl = "https://www.bbc.com/news"
res = R.get(targetUrl)
html = res.text
bs = BS.BeautifulSoup(html)
#print(bs.prettify())

# Get all new link 
linkTags = bs.find_all("a")
candidateLinks=[]
for linkTag in linkTags:
    href = linkTag["href"]
    if href is not None:
        if href.startswith("/news/"):
            candidateLinks.append("https://www.bbc.com"+href)
print(candidateLinks[:10])

# Retrieve data from the candidate links
for link in candidateLinks[:20]:
    res = R.get(link)
    bs = BS.BeautifulSoup(res.text)
    # display information
    print(bs.title.text)
    paras = bs.find_all("p")
    text = [para.text for para in paras]
    print(text)
    print("-------------------------------------------")
    
```

```python
#1
print("1.")
strs = ["ABCDEFabcdef123450","ABCD@Fabcdef123450"]
ans=[]
for text in strs:
    y= re.fullmatch(r"[a-zA-Z0-9]+", text)
    if y is not None:
        ans.append(y.group())
print(strs)
print(ans)
print(" ")

#2
print("2.")
strs = ['bab', 'abbbbb','baaaa']
ans=[]
for text in strs:
    y= re.search(r"\w*ab+\w*", text)
    if y is not None:
        ans.append(y.group())
print(strs)
print(ans) 
print(" ")

#3
print("3.")
strs = ['65117896', '78238936', '56666665']
ans=[]
for text in strs:
    y= re.search(r"^6\d*6", text)
    if y is not None:
        ans.append(y.group())
print(strs)
print(ans)
print(" ")

#4
print("4.")
strs = ['1', '23', '345', '45678']
ans=[]
for text in strs:
    y= re.search(r"\d{2,4}", text)
    if y is not None:
        ans.append(y.group())
print(strs)
print(ans) 
print(" ")

#5
print("5.")
strs = ['210.08.090.194', '010.01.010.100']
ans=[]
for text in strs:
    y= re.sub(r"^0+","",text)
    if y is not None:
        ans.append(y)
print(strs)
print(ans) 
print(" ")

#6
print("6.")
strs = ['Python Exercises Of Regular_Expression']
ans=[]
for text in strs:
    y= re.sub(r"( |_)",lambda m :"_" if m.group() == ' ' else " ",text)
    if y is not None:
        ans.append(y)
print(strs)
print(ans) 
print(" ")

#7
print("7.")
strs = ['2022-09-10']
ans=[]
for text in strs:
    y= re.sub(r"(\d{4})-(\d{2})-(\d{2})",r"\3-\2-\1",text)
    if y is not None:
        ans.append(y)
print(strs)
print(ans) 
print(" ")

#8
print("8.")
strs = ['The following example creates an ArrayList with a capacity of 50 elements. Four elements are then added to the ArrayList and the ArrayList is trimmed accordingly.']
ans=[]
for text in strs:
    y= re.findall(r"\b[ae]\w+", text)
    if y is not None:
        ans.append(y)
print(strs)
print(ans) 
print(" ")

#9
print("9.")
strs = ['Regex can be used in programming languages such as “Python”, “SQL”, “Javascript”, “R”, “GoogleAnalytics”, “Google Data Studio”, and throughout the coding process.']
ans=[]
for text in strs:
    y= re.findall(r"\“([\w\s]*)\”", text)
    if y is not None:
        ans.append(y)
print(strs)
print(ans) 
print(" ")

#10
print("10.")
strs = ['Find more Examples at Github https://www.github.com or W3School https://www.w3schools.com/.']
ans=[]
for text in strs:
    y= re.findall(r"http[\w://.]*.com", text)
    if y is not None:
        ans.append(y)
print(strs)
print(ans) 
print(" ")

```

```python
# import lib
import xml.etree.ElementTree as ET
from io import open
import re
# load xml file 
def Xml_To_Text(path:str)->str:
    tree:ET= ET.parse(path)
    root:ET.Element = tree.getroot()
    text:str = ET.tostring(root, encoding='utf-8', method='text').decode('utf-8')
    y= re.sub(r"( ){2,}","",text)
    y= re.sub(r"\n+","\n",y)
    text = y
    return text

xmlPath:str = "./sample-src.xml"
res = Xml_To_Text(xmlPath)
print(res)
with open('./OutputFile/FromXMLText.txt', 'w', encoding="utf-8") as file:
    # Write the text content to the file
    file.write(res)
```
